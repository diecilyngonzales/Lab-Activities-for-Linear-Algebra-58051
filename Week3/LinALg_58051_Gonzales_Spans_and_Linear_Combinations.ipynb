Task 1 : 

Let:

$$
  x_1=\begin{bmatrix} 3 \\ 2 \end{bmatrix},
  x_2=\begin{bmatrix} 2 \\ -4 \end{bmatrix},
  x_3=\begin{bmatrix} -3 \\ -5 \end{bmatrix},
  c_1=5,c_2=-4,c_3=2
$$

Find and graph:

$$X_1=c_1*x_1,$$
$$X_2=c_2*x_2,$$
$$X_3=c_3*x_3,$$

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

vectX1 = np.array([3,2])
vectX2 = np.array([2,-4])
vectX3 = np.array([-3,-5])

c1 = 5
c2 = -4
c3 = 2

A = vectX1*c1
B = vectX2*c2
C = vectX3*c3

plt.quiver(0,0,vectX1[0],vectX1[1],angles='xy',scale_units='xy',scale=1,label="X1",color='red')
plt.quiver(0,0,vectX2[0],vectX2[1],angles='xy',scale_units='xy',scale=1,label="X2",color='yellow')
plt.quiver(0,0,vectX3[0],vectX3[1],angles='xy',scale_units='xy',scale=1,label="X3",color='green')

plt.quiver(0,0,A[0],A[1],angles='xy',scale_units='xy',scale=1,label="c1*X1",color='maroon')
plt.quiver(0,0,B[0],B[1],angles='xy',scale_units='xy',scale=1,label="c2*X2",color='orange')
plt.quiver(0,0,C[0],C[1],angles='xy',scale_units='xy',scale=1,label="c3*X3",color='darkgreen')

plt.xlim(-20,20)
plt.ylim(-20,20)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.show()
```

Task 2 : 

Let:

$$
  x_1=\begin{bmatrix} 3 \\ 2 \end{bmatrix},
  x_2=\begin{bmatrix} 8 \\ -4 \end{bmatrix},
  x_3=\begin{bmatrix} 0 \\ -7 \end{bmatrix},
  x_4=\begin{bmatrix} -5 \\ 6 \end{bmatrix},
  x_5=\begin{bmatrix} -9 \\ -4 \end{bmatrix},
  x_6=\begin{bmatrix} 3 \\ 5 \end{bmatrix}
$$

Find and graph:
$$R_1 = x_1+x_4$$
$$R_2 = x_2+x_6$$
$$R_3 = x_3+x_5+x_1$$

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
 
x1 = np.array([3,2])
x2 = np.array([8,-4])
x3 = np.array([0,-7])
x4 = np.array([-5,6])
x5 = np.array([-9,-4])
x6 = np.array([3,5])
 
R1 = x1 + x4
R2 = x2 + x6
R3 = x3 + x5 + x1
 
plt.quiver(0,0,x1[0],x1[1],angles='xy',scale_units='xy',scale=1,label='x1',color='blue')
plt.quiver(x1[0],x1[1],x4[0],x4[1],angles='xy',scale_units='xy',scale=1,label='x4',color='blue')
plt.quiver(0,0,R1[0],R1[1],angles='xy',scale_units='xy',scale=1,label='R1',color='red')
 
plt.quiver(0,0,x2[0],x2[1],angles='xy',scale_units='xy',scale=1,label='x2',color='yellow')
plt.quiver(x2[0],x2[1],x6[0],x6[1],angles='xy',scale_units='xy',scale=1,label='x6',color='yellow')
plt.quiver(0,0,R2[0],R2[1],angles='xy',scale_units='xy',scale=1,label='R2',color='red')
 
plt.quiver(0,0,x3[0],x3[1],angles='xy',scale_units='xy',scale=1,label='x3',color='green')
plt.quiver(x3[0],x3[1],x5[0],x5[1],angles='xy',scale_units='xy',scale=1,label='x5',color='green')
plt.quiver(x3[0]+x5[0],x3[1]+x5[1],x1[0],x1[1],angles='xy',scale_units='xy',scale=1,label='x1',color='green')
plt.quiver(0,0,R3[0],R3[1],angles='xy',scale_units='xy',scale=1,label='R3',color='red')
 
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.show()
```

2D Vector Visualization

```
 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
 
vectA = np.array([1,2])
vectB = np.array([3,-2])
R = np.arange(-9,9,1)
 
c1,c2 = np.meshgrid(R,R)
vectR = vectA + vectB 
spanRx = c1*vectA[0] + c2*vectB[0]
spanRy = c1*vectA[1] + c2*vectB[1]
 
plt.scatter(spanRx,spanRy, s=3, alpha=0.95)
 
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.show()
```

3D Vector Visualization

```
 
import numpy as np
import matplotlib.pyplot as plt
 
vectX = np.array([3,6,2])
vectY = np.array([8,6,-3])
vectZ = np.array([5,12,2])
 
R = np.arange(-10,10,2)
 
c1,c2,c3 = np.meshgrid(R,R,R)
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') 
 
spanX = c1*vectX[0]+c2*vectY[0]+c3*vectZ[0]
spanY = c1*vectX[1]+c2*vectY[1]+c3*vectZ[1] 
spanZ = c1*vectX[2]+c2*vectY[2]+c3*vectZ[2]
 
ax.scatter(spanX,spanY,spanZ,s=5,alpha=0.65)
 
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis') 
 
 
plt.show()
```

4D Vector Visualization

```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm
 
vectX = np.array([5,-4,3,-4])
vectY = np.array([7,2,9,2])
vectZ = np.array([-8,1,5,4])
vectW = np.array([-3,5,0,-5])
 
R = np.arange(-10,10,2)
 
c1,c2,c3,c4 = np.meshgrid(R,R,R,R)
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
cmap = plt.cm.get_cmap('viridis') 
norm = matplotlib.colors.Normalize(vmin=-100, vmax=100)
sm = matplotlib.cm.ScalarMappable(cmap=cmap, norm=norm)
 
spanX = c1*vectX[0]+c2*vectY[0]+c3*vectZ[0]+c4*vectW[0] 
spanY = c1*vectX[1]+c2*vectY[1]+c3*vectZ[1]+c4*vectW[1] 
spanZ = c1*vectX[2]+c2*vectY[2]+c3*vectZ[2]+c4*vectW[2]
spanW = c1*vectX[3]+c2*vectY[3]+c3*vectZ[3]+c4*vectW[3]
 
sm.set_array(spanW)
 
ax.scatter(spanX,spanY,spanZ,s=5,c=spanW)
 
cbar = plt.colorbar(sm)
 
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
 
plt.show()
```
