Let:

$$
  x = \begin{bmatrix} 3 \\ 4 \\ 2\end{bmatrix},
  y = \begin{bmatrix} -2 \\ 5 \\ 7\end{bmatrix} 
$$

Then:
$$
  A = \begin{bmatrix} 3 \ -2\\4\ 5 \\ 2\ 7\end{bmatrix},
$$
or: 
$$
  A \left\{ \begin{array} 3x -2y \\ 4x +5y \\ 2x + 7y\end{array}\right.
$$


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

A = np.array([
              [3,-2],
              [4,5],
              [2,7]
              ])

plt.xlim(0,15)
plt.ylim(-15,15)
plt.axhline(y=0, color='grey')
plt.grid()

c = np.arange(0,100)

plt.plot(c*A[0,0], c*A[0,1], color='red')
plt.plot(c*A[1,0], c*A[1,1], color='blue')
plt.plot(c*A[2,0], c*A[2,1], color='yellow')

plt.title('''2-Dimensional System of Linear Equations
A = {3x - 2y,
      4x + 5y,
      2x + 7y
''')
plt.show()
```

Let:

$$
  x = \begin{bmatrix} 2 \\ 4 \\ 7 \end{bmatrix},
  y = \begin{bmatrix} 1 \\ 9 \\ 3 \end{bmatrix},
  z = \begin{bmatrix} 2 \\ 5 \\ 6 \end{bmatrix}
$$

Then:

$$A = \begin{bmatrix} 2 \ 1 \ 2 \\ 4 \ 9 \ 5 \\ 7 \ 3 \ 6 \end{bmatrix},$$
or:
$$
A = \left\{\begin{array} 2x + y + 2z \\ 4x + 9y + 5z \\ 7x +3y +6z \end{array}\right.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlim([0,10])
ax.set_ylim([0,10])
ax.set_zlim([0,10])

A = np.array([[2,1,2],
              [4,9,5],
              [7,3,6]])

c = np.arange(0,3)

colors = ['g', 'r', 'b']
for i in range(A.shape[0]):
    ax.plot(c*A[i,0], c*A[i,1], c*A[i,2], color=colors[i])

plt.title('''3-Dimensional System of Linear Equations
A = {2x + y + 2z,
      4x + 9y + 5z,
      7x + 3y + 6z
''')
plt.show()
```
